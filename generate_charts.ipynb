```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Employee Performance
emp_df = pd.read_csv('MVP_employee.csv')
emp_df = emp_df.dropna(subset=['total_invoice']).copy()
emp_df['total_invoice'] = pd.to_numeric(emp_df['total_invoice'], errors='coerce')
emp_df = emp_df.dropna(subset=['total_invoice']).sort_values('total_invoice', ascending=False)

plt.figure(figsize=(5, 3))
plt.barh(emp_df['first_name'] + ' ' + emp_df['last_name'], emp_df['total_invoice'], color='#2A9D8F')
plt.title('Sales Support Representatives by Invoice Total')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('Visualizations/employee_performance.png', bbox_inches='tight', dpi=150)
plt.show()
```

```python
# 2. Global Revenue
country_df = pd.read_csv('countries_generating_money.csv')
top_5_countries = country_df.nlargest(5, 'total_revenue').sort_values('total_revenue', ascending=False)
all_countries = country_df.sort_values('total_revenue', ascending=False)
other_revenue = country_df.loc[~country_df['country'].isin(top_5_countries['country']), 'total_revenue'].sum()
pie_countries = pd.concat([
	top_5_countries[['country', 'total_revenue']],
	pd.DataFrame([{'country': 'Other countries', 'total_revenue': other_revenue}])
], ignore_index=True)

# Pie Chart
plt.figure(figsize=(5, 3))
plt.pie(pie_countries['total_revenue'], labels=pie_countries['country'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Global Share (Top 5 Countries + Others)')
plt.tight_layout()
plt.savefig('Visualizations/top_5_countries_pie.png', bbox_inches='tight', dpi=150)
plt.show()

# Bar Chart
plt.figure(figsize=(5, 3))
plt.barh(top_5_countries['country'], top_5_countries['total_revenue'], color='#457B9D')
plt.title('Top 5 Countries by Revenue')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('Visualizations/top_5_countries_bar.png', bbox_inches='tight', dpi=150)
plt.show()
```

```python
# 3. Top Genres Sold
genres_df = pd.read_csv('Genres.csv').sort_values('quantity', ascending=False)

plt.figure(figsize=(6, 3))
plt.barh(genres_df['name'], genres_df['quantity'], color='#E9C46A')
plt.title('Top 5 Genres Sold')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('Visualizations/top_5_genres_bar.png', bbox_inches='tight', dpi=150)
plt.show()
```

```python
# 4. VIP Customers
customers_df = pd.read_csv('top_5_customers.csv').sort_values('total_spent', ascending=False)

plt.figure(figsize=(6, 3))
plt.barh(customers_df['first_name'] + ' ' + customers_df['last_name'], customers_df['total_spent'], color='#E76F51')
plt.title('Top 5 Highest-Spending Customers')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('Visualizations/top_5_customers_bar.png', bbox_inches='tight', dpi=150)
plt.show()
```

```python
# 5. Top Rock Cities
rock_cities_df = pd.read_csv('top_10_rock_cities.csv').sort_values('num_rock_funs', ascending=False)

# Bar Chart
plt.figure(figsize=(6, 3))
plt.barh(rock_cities_df['billing_city'], rock_cities_df['num_rock_funs'], color='#6D597A')
plt.title('Top 10 Rock Cities')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('Visualizations/top_10_rock_cities_bar.png', bbox_inches='tight', dpi=150)
plt.show()
```
