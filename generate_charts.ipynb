```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
employee_df = pd.read_csv('MVP_employee.csv')
employee_df['employee_name'] = employee_df['first_name'] + ' ' + employee_df['last_name']

plt.figure(figsize=(9, 4))
plt.barh(employee_df['employee_name'], employee_df['total_invoice'], color='steelblue')
plt.xlabel('Total Sales ($)')
plt.ylabel('Employee')
plt.title('Total Invoice by Employee')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('visualizations/employee_performance.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
country_df = pd.read_csv('countries_generating_money.csv')
country_df['total_revenue'] = pd.to_numeric(country_df['total_revenue'])

# Show the top 5 countries as slices and group the rest into a single category.
top_5_countries = country_df.nlargest(5, 'total_revenue').copy()
other_revenue = country_df['total_revenue'].sum() - top_5_countries['total_revenue'].sum()

pie_labels = top_5_countries['country'].tolist() + ['Other Countries']
pie_values = top_5_countries['total_revenue'].tolist() + [other_revenue]

plt.figure(figsize=(5, 4))
plt.pie(
    pie_values,
    labels=pie_labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=plt.cm.Set3.colors,
    wedgeprops={'edgecolor': 'white'}
)
plt.title('Top 5 Countries by Total Revenue')
plt.axis('equal')
plt.tight_layout()
plt.savefig('visualizations/top_5_countries_pie.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
country_df = pd.read_csv('countries_generating_money.csv')
country_df['total_revenue'] = pd.to_numeric(country_df['total_revenue'])

# Horizontal bar chart for the top 5 countries by revenue.
top_5_countries = country_df.nlargest(5, 'total_revenue').sort_values('total_revenue', ascending=True)

plt.figure(figsize=(6, 3))
plt.barh(top_5_countries['country'], top_5_countries['total_revenue'], color='coral')
plt.xlabel('Total Revenue ($)')
plt.ylabel('Country')
plt.title('Top 5 Countries by Total Revenue')
plt.tight_layout()
plt.savefig('visualizations/top_5_countries_bar.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
genres_df = pd.read_csv('Genres.csv')
genres_df['quantity'] = pd.to_numeric(genres_df['quantity'])

top_5_genres = genres_df.nlargest(5, 'quantity').sort_values('quantity', ascending=True)

plt.figure(figsize=(10, 5))
plt.barh(top_5_genres['name'], top_5_genres['quantity'], color='mediumseagreen')
plt.xlabel('Quantity Sold')
plt.ylabel('Genre')
plt.title('Top 5 Genres by Quantity Sold')
plt.tight_layout()
plt.savefig('visualizations/top_5_genres_bar.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
customers_df = pd.read_csv('top_5_customers.csv')

top_5_customers = customers_df.nlargest(5, 'total_spent').sort_values('total_spent', ascending=True)


plt.figure(figsize=(10, 5))
plt.barh(top_5_customers['first_name'] + ' ' + top_5_customers['last_name'], top_5_customers['total_spent'], color='mediumpurple')
plt.xlabel('Total Spent')
plt.ylabel('Customer')
plt.title('Top 5 VIP Customers by Total Spending')
plt.tight_layout()
plt.savefig('visualizations/top_5_customers_bar.png', bbox_inches='tight', dpi=300)
plt.show()
```

```python
top_10_rock_cities_df = pd.read_csv('top_10_rock_cities.csv')

top_10_rock_cities_df = top_10_rock_cities_df.nlargest(10, 'num_rock_funs').sort_values('num_rock_funs', ascending=True)

plt.figure(figsize=(10, 5))
plt.barh(top_10_rock_cities_df['billing_city'], top_10_rock_cities_df['num_rock_funs'], color='darkorange')
plt.xlabel('Number of Rock Fans')
plt.ylabel('City')
plt.title('Top 10 Cities with Most Rock Fans')
plt.tight_layout()
plt.savefig('visualizations/top_10_rock_cities_bar.png', bbox_inches='tight', dpi=300)
plt.show()
```
